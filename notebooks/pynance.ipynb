```python
import pynance as pn
import matplotlib.pyplot as plt

# Download historical data for a ticker (e.g., 'AAPL')
df = pn.data.get('AAPL', start='2023-01-01', end='2024-01-01')

# Calculate 20-day moving average
df['SMA_20'] = df['Close'].rolling(window=20).mean() # Calculate 20-day moving average, window=20 is the number of days

# Calculate daily returns
df['Return'] = df['Close'].pct_change()

# Calculate annual volatility
annual_volatility = df['Return'].std() * (252 ** 0.5) # Annualize the volatility, assuming 252 trading days in a year
print(f"Annual Volatility: {annual_volatility}")

# Plot Close price and SMA
plt.figure(figsize=(14, 7))
plt.plot(df['Close'], label='Close Price')
plt.plot(df['SMA_20'], label='20-Day SMA', color='orange')
plt.title('Close Price and 20-Day SMA')
plt.legend()
plt.show()
```
